```python
from gliner import GLiNER
# 1. Define our new text

# 3. Load the PII model
model = GLiNER.from_pretrained("nvidia/gliner-pii")
```

```python
text = """
14 Other
12b
12c
1234 S MAPLE ST
12d
ANYWHERE, CA 123456
13 Stat emp| Ret. plan 3rd party sick pay
ADP.
15 StateEmployer's state ID no 16 State wages, tips, etc.
Always Designing
CA 12345678901ABC
23500.00
for People
17 State income tax
"""

# 2. Define the labels we're hunting for.
labels = [    "full_name", "first_name", "last_name",
    "email", "phone_number",
    "date_of_birth",
    "passport_number",
    "national_id", "ssn",
    "street_address", "address", "mailing_address",
    "credit_card", "bank_account_number",
    "ip_address", "city", "state", "postcode", "po_box"
]

# 4. Run the prediction at given threshold
entities = model.predict_entities(text, labels, threshold=0.1)

for e in entities:
  print(e)
```
